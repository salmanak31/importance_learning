```python
import sys
sys.path.append('../')
%matplotlib notebook
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

import il_pedagogical as ilp
```

```python
A = 3264.7
B = 95.373*5
l1 = 3.2394 * 2.33
l2 = 1.35258 * 2.33
l3 = l2


# A = 3164.7/100
# B = 900.373/100

# l1 = 5.25
# l2 = 0.9
# l3 = l2

beta = 0.33975
n = 22.956
# n = 12.956

def energy(coords, p):
    x, y = coords
    return sum([A*np.exp(-l1 * dist(x, p[i, 0], y, p[i, 1])) - BO(x, y, i, p, beta, l3, n) * B *np.exp(-l2 * dist(x, p[i, 0], y, p[i, 1])) for i in range(len(p))])

def BO(x, y, i, p, beta, l3, n):
    thing = [np.exp(l3**3 * (dist(x, p[i, 0], y, p[i, 1]) - dist(x, p[k, 0], y, p[k, 1]))**3) for k in range(len(p)) if k !=i ]
    return (1 + beta**n * np.sum(thing))**(-0.5 * n)
    
def dist(xm, xi, ym, yi):
    return np.sqrt((xm - xi)**2 + (ym - yi)**2)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5,4))
lattice_size = 20
a = 35 # ~ Potential width, dimensionless
D = 60 # Potential depth, kJ/mol
r_eq = 1.0 # Equilibrium distance of unperturbed lattice, dimensionless
T = 300
delta = 0.25
gauss_pot_std = 0.9
dissociation_type = 2
potential_type = 2

displacement_type = 'uniform'
lattice = ilp.make_quenched_disorder_lattice(lattice_size, delta, displacement_type, False, seed=True)
ilp.plot_lattice(lattice)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7,6))
p = lattice[4:6, 4:6].reshape(4,2)
# p = lattice[2, 2].reshape(1,2)

lower = 3
upper = 6

xaxis = np.linspace(lower, upper, 100)
yaxis = np.linspace(lower, upper, 100)
x, y = np.meshgrid(xaxis, yaxis)

energy_landscape = energy((x, y), p)
initial_guess = np.asarray([4.5, 4.5])
result = minimize(energy, initial_guess, args=p)

plt.contourf(energy_landscape, extent=[lower, upper, lower, upper], levels=15)
plt.colorbar()

plt.scatter(initial_guess[0], initial_guess[1], s=100, c="white", marker='x', label="initial guess")
plt.scatter(result.x[0], result.x[1], s=100, c="red", marker='x', label="minimum")
plt.legend()
plt.show()

print('Optmized gemoetry: ({:03.3f}, {:03.3f})'.format(result.x[0], result.x[1]))
```
